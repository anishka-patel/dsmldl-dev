```python
import os
import pickle
import pandas as pd
```

```python
SAMPLE_DIR = '../input/ieee-fraud-detection/sample_submission.csv'
```

```python
INPUT_DIR = '.'
data = {}
files = os.listdir(INPUT_DIR)
for file in files:
    if 'Xtrain' in file:
        if 'pca_50' in file:
            data['Xtrain_pca_50'] = INPUT_DIR + '/' + file
        elif 'pca_100' in file:
            data['Xtrain_pca_100'] = INPUT_DIR + '/' + file
        else:
            data['Xtrain'] = INPUT_DIR + '/' + file
    if 'Ytrain' in file:
        data['Ytrain'] = INPUT_DIR + '/' + file
    if 'Xtest' in file:
        if 'pca_50' in file:
            data['Xtest_pca_50'] = INPUT_DIR + '/' + file
        elif 'pca_100' in file:
            data['Xtest_pca_100'] = INPUT_DIR + '/' + file
        else:
            data['Xtest'] = INPUT_DIR + '/' + file
        
data
```

```python
Xtest = pickle.load(open(data['Xtest_pca_50'], 'rb'))
Xtest.head()
```

```python
clf = pickle.load(open('./pca_50_models/model_2.model', 'rb'))
```

```python
Ytest_ = clf.predict_proba(Xtest)
Ytest_
```

```python
Y_ = []
for el in Ytest_:
    Y_.append(el[1])
Y_
```

```python
submission = pd.read_csv(SAMPLE_DIR)
submission.head()
```

```python
submission['isFraud'] = Y_
submission.head()
```

```python
submission.to_csv('prediction_pca_50_model_2.csv', index=False)
```
