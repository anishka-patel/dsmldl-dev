```python
def check_null_values(df):
    for column in df.columns:
        print('Column Name:', column, '\nHas null values:',df[column].isna().any(), '\nAmount of null values:', df[column].isna().sum())
        print('Values - Count')
        print(df[column].value_counts())
        print()
```

```python
def replace_null_values(df):
    for column in df.columns:
        if(df[column].isna().any()):
            print(column, df[column].isna().sum())
            def replace():
                global i
                values = df[column].value_counts().index
                counts = df[column].value_counts().values
                counts = counts/counts.sum()
                yield np.random.choice(values,replace=False, p=counts)
            new_value = replace()
            df[column].fillna(value=next(new_value), inplace=True)
    for column in df.columns:
        print(column, df[column].isna().sum())
        if(df[column].isna().any()):
            print('Column:', column, 'has null values')
        else:
            print('No Null values in column')
```

```python
df_train = df_train_tid.join(df_train_id, on='TransactionID',how='left', sort=True)
df_test = df_test_tid.join(df_test_id, on='TransactionID', how='left', sort=True)
```

```python
del(df_train_tid)
del(df_train_id)
del(df_test_tid)
del(df_test_id)
```

```python
df_train['isFraud'].fillna(value=0.0, inplace=True)
df_test['isFraud'] = -1
```

```python
def mod(x):
    if type(x) == type(''):
        x = x.split(sep='x')
        x = int(x[0])*int(x[1])
        return x
    else:
        return np.nan
```

```python
df_X['id_33'] = df_X['id_33'].map(mod)
df_X['id_33'].value_counts()
```

```python
dum_cols = []
for column in df_X.columns:
    if df_X[column].dtype == 'object':
        dum_cols.append(column)
dum_cols
```

```python
df_X = pd.get_dummies(df_X, columns=dum_cols)
print(df_X.shape)
del(dum_cols)
```

```python
df_train = df_X[df_X['isFraud'] != -1]
df_test = df_X[df_X['isFraud'] == -1]
del(df_X)
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
Ytrain = df_train['isFraud']
Xtrain = df_train.drop(columns=['isFraud'])
Xtest = df_test.drop(columns=['isFraud'])
print(Xtrain.shape)
print(Ytrain.shape)
print(Xtest.shape)
pickle.dump(Ytrain, open('Ytrain_ANN.data', 'wb'))
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(Xtrain)
pickle.dump(scaler, open('scaler_ANN.scaler', 'wb'))
Xtraincols = Xtrain.columns
Xtrainindex = Xtrain.index
Xtestcols = Xtest.columns
Xtestindex = Xtest.index
```

```python
Xtrain = pd.DataFrame(data=scaler.transform(Xtrain), index=Xtrainindex, columns=Xtraincols)
```

```python
pickle.dump(Xtrain, open('Xtrain_ANN.data', 'wb'), protocol=4)
```

```python
Xtest = pd.DataFrame(data=scaler.transform(Xtest), index=Xtestindex, columns=Xtestcols)
```

```python
pickle.dump(Xtest, open('Xtest_ANN.data', 'wb'), protocol=4)
```

```python
del(scaler)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
from keras import optimizers
```

```python
clf = Sequential()
```

```python
clf.add(Dense(units=400, activation='relu', kernel_initializer='uniform', input_dim=598, bias_initializer='zeros'))
clf.add(Dropout(rate=0.1))
```

```python
clf.add(Dense(units=400, activation='relu', kernel_initializer='uniform', bias_initializer='zeros'))
clf.add(Dropout(rate=0.1))
```

```python
clf.add(Dense(units=400, activation='relu', kernel_initializer='uniform', bias_initializer='zeros'))
clf.add(Dropout(rate=0.1))
```

```python
clf.add(Dense(units=1, activation='sigmoid', kernel_initializer='uniform'))
```

```python
opt_adam = optimizers.Adam(lr=0.001, epsilon=1e-8)
```

```python
clf.compile(optimizer=opt_adam, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
clf.fit(Xtrain, Ytrain, epochs=50, validation_split=0.1, shuffle=True)
```

```python
clf.evaluate(Xtrain,Ytrain)
```

```python
pickle.dump(clf, open('clf_ANN.model', 'wb'))
clf.save('clf_ANN.h5')
clf.summary()
```

```python
from sklearn.metrics import log_loss, classification_report, confusion_matrix
```

```python
Ytrain_ = clf.predict(Xtrain) > 0.5
```

```python
print(log_loss(Ytrain, Ytrain_))
print(confusion_matrix(Ytrain, Ytrain_))
print(classification_report(Ytrain, Ytrain_))
```
