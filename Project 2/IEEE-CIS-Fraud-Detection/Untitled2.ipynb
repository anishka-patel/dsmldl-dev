```python
import os
import pickle
import pandas as pd
```

```python
SAMPLE_DIR = '../input/ieee-fraud-detection/sample_submission.csv'
```

```python
INPUT_DIR = '../input/ieee-fraud-detection-dataprocessing'
data = {}
files = os.listdir(INPUT_DIR)
for file in files:
    if 'Xtrain' in file:
        data['Xtrain'] = INPUT_DIR + '/' + file
    if 'Ytrain' in file:
        data['Ytrain'] = INPUT_DIR + '/' + file
    if 'Xtest' in file:
        data['Xtest'] = INPUT_DIR + '/' + file
data
```

```python
Xtest = pickle.load(open(data['Xtest'], 'rb'))
Xtest.head()
```

```python
clf = pickle.load(open('./models/model_9.model', 'rb'))
```

```python
Ytest_ = clf.predict_proba(Xtest)
Ytest_
```

```python
Y_ = []
for el in Ytest_:
    Y_.append(el[1])
Y_
```

```python
submission = pd.read_csv(SAMPLE_DIR)
submission.head()
```

```python
submission['isFraud'] = Y_
submission.head()
```

```python
submission.to_csv('prediction_model_9.csv', index=False)
```
